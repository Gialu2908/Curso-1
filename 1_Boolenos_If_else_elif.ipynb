## **Escritura de instrucciones "if"**

Para expresar la lgica condicional en Python, se usan instrucciones if. Al escribir una instruccin if, se basa en otro concepto que se describe en este mdulo, el de los operadores matemticos. Python admite los operadores lgicos comunes de matemticas: igual, no igual, menor que, menor o igual que, mayor que y mayor o igual que. Probablemente est acostumbrado a ver que estos operadores se muestran mediante smbolos, que tambin es la forma en que se representan en Python.

*  Es igual que: a == b
*  No es igual a: a != b
*  Menor que: a < b
*  Menor o igual que: a <= b
*  Mayor que: a > b
*  Mayor o igual que: a >= b

# Expresiones de prueba
Debe usar una instruccin if para ejecutar cdigo solo si se cumple una condicin concreta. Lo primero que hace al escribir una instruccin if es comprobar la condicin mediante una expresin de prueba. Despus, se determina si la instruccin se evala como True o False. Si es True, se ejecuta el siguiente bloque de cdigo con sangra:



```
a = 97
b = 55
# test expression
if a < b:
    # statement to be run
    print(b)
```



En este ejemplo, a < b es la expresin de prueba. El programa evala la expresin de prueba y, despus, ejecuta el cdigo dentro de la instruccin if solo si la expresin de prueba es True. Si evala la expresin, sabe que es False, por lo que no se ejecutar ningn cdigo que escriba en la instruccin if.

 ***Nota***

En Python, None y 0 tambin se interpretan como False.

# **Escritura de instrucciones if**
Use una instruccin if si quiere ejecutar cdigo solo si se cumple una condicin concreta. La sintaxis de una instruccin if siempre es la siguiente:

````python
if test_expression:
    # sentencia(s) a ejecutar
````

````python
a = 93
b = 27
if a >= b:
    print(a)
````

En Python, se debe aplicar sangra al cuerpo de una instruccin if. Siempre se ejecutar cualquier cdigo que siga a una expresin de prueba que no tenga sangra:

````python
a = 24
b = 44
if a <= 0:
    print(a)
print(b)
````

En este ejemplo, la salida es 44 ya que la expresin de prueba es False y la instruccin print(b) no tiene sangra en el mismo nivel que la instruccin if.

# **Qu son las instrucciones "else" y "elif"?**

Qu ocurre si tambin quiere que el programa ejecute un fragmento de cdigo cuando la expresin de prueba es False? O bien, qu ocurre si quiere incluir otra expresin de prueba? Python tiene otras palabras clave que puede usar para crear instrucciones if ms complejas, else y elif. Al usar if, else y elif de forma conjunta, puede escribir programas complejos con varias expresiones de prueba e instrucciones para ejecutar.

# **Uso de else**

Sabe que cuando se usa una instruccin if, el cuerpo del programa solo se ejecutar si la expresin de prueba es True. Para agregar ms cdigo que se ejecute cuando la expresin de prueba sea False, debe agregar una instruccin else.

Ahora se volver al ejemplo de la seccin anterior:

````python
a = 93
b = 27
if a >= b:
    print(a)
````

En este ejemplo, si a no es mayor o igual que b, no ocurre nada. Imagine que en su lugar quiere imprimir b si la expresin de prueba es False:

````python
a = 93
b = 27
if a >= b:
    print(a)
else:
    print(b)
````

Si la expresin de prueba es False, se omite el cdigo del cuerpo de la instruccin if y el programa contina ejecutndose desde la instruccin else. La sintaxis de una instruccin if/else siempre es la siguiente:



```
if test_expression:
    # sentencia(s) a ejecutar
else:
    # sentencia(s) a ejecutar
```



# **Uso de elif**

En Python, la palabra clave elif es la abreviatura de else if. El uso de instrucciones elif permite agregar varias expresiones de prueba al programa. Estas instrucciones se ejecutan en el orden en que se escriben, por lo que el programa escribir una instruccin elif solo si la primera instruccin if es False. Por ejemplo:

````python
a = 93
b = 27
if a >= b:
    print("a es mayor o igual que b")
elif a == b:
    print("a is equal to b")
````

La instruccin elif de este bloque de cdigo no se ejecutar, porque la instruccin if es True.

La sintaxis de una instruccin if/elif siempre es la siguiente:



```
if test_expression:
    # sentencia(s) a ejecutar
elif test_expression:
    # sentencia(s) a ejecutar
```



# **Combinacin de instrucciones if, elif y else**

Puede combinar instrucciones if, elif y else para crear programas con lgica condicional compleja. Recuerde que una instruccin elif solo se ejecuta cuando la condicin if es false. Tenga en cuenta tambin que un bloque if solo puede tener un bloque else, pero puede tener varios bloques elif.

Ahora se volver a examinar el ejemplo con una instruccin elif agregada:

````python
a = 93
b = 27
if a > b:
    print("a es mayor que b")
elif a < b:
    print("a es menor que b")
else: 
    print ("a es igual a b")
````

Un bloque de cdigo que usa los tres tipos de instrucciones tiene la sintaxis siguiente:



```
if test_expression:
    # sentencia(s) a ejecutar
elif test_expression:
    # sentencia(s) a ejecutar
elif test_expression:
    # sentencia(s) a ejecutar
else:
    # sentencia(s) a ejecutar
```



# **Uso de lgica condicional anidada**
Python tambin admite la lgica condicional anidada, lo que significa que puede anidar instrucciones if, elif y else para crear programas an ms complejos. Para anidar condiciones, aplique sangra a las condiciones internas y todo lo que est en el mismo nivel de sangra se ejecutar en el mismo bloque de cdigo:

````python
a = 16
b = 25
c = 27
if a > b:
    if b > c:
        print ("a es mayor que b y b es mayor que c")
    else: 
        print ("a es mayor que b y menor que c")
elif a == b:
    print ("a es igual a b")
else:
    print ("a es menor que b")
````

Este fragmento de cdigo genera la salida a is less than b.

La lgica condicional anidada sigue las mismas reglas que la lgica condicional convencional dentro de cada bloque de cdigo. Este es un ejemplo de la sintaxis:



```
if test_expression:
    # sentencia(s) a ejecutar
    if test_expression:
        # sentencia(s) a ejecutar
    else: 
        # sentencia(s) a ejecutar
elif test_expression:
    # sentencia(s) a ejecutar
    if test_expression:
        # sentencia(s) a ejecutar
    else: 
        # sentencia(s) a ejecutar
else:
    # sentencia(s) a ejecutar
```

# **Qu son los operadores "and" y "or"?**

En ocasiones, es posible que quiera combinar expresiones de prueba para evaluar varias condiciones en una instruccin if, elif o else. En este caso, usara los operadores booleanos and y or.

# **El operador "or"**
Puede conectar dos expresiones booleanas, o de prueba, mediante el operador booleano or. Para que toda la expresin se evale como True, al menos una de las subexpresiones debe ser true. Si ninguna de las subexpresiones es true, toda la expresin se evala como False. Por ejemplo, en la expresin siguiente, toda la expresin de prueba se evala como True, porque se ha cumplido una de las condiciones de las subexpresiones:

````python
a = 23
b = 34
if a == 34 or b == 34:
    print(a + b)
````

Si las dos subexpresiones son true, toda la expresin de prueba tambin se evala como True.

Una expresin booleana que usa or tiene la sintaxis siguiente:



```
sub-expression1 or sub-expression2
```



# **El operador "and"**
Tambin puede conectar dos expresiones de prueba mediante el operador booleano and. Las dos condiciones de la expresin de prueba deben cumplirse para que toda la expresin de prueba se evale como True. En cualquier otro caso, la expresin de prueba es False. En el ejemplo siguiente, toda la expresin de prueba se evala como False, porque solo una de las condiciones de las subexpresiones es true:

````python
a = 23
b = 34
if a == 34 and b == 34:
    print (a + b)
````
